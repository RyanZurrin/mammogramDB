```python
omama_loader = O.OmamaLoader()
data = O.Data(data_loader=omama_loader, timing = True, print_data= True, load_cache=True)
```

```python
# filter by labels and by minimum row and col size
data.filter_data(labels=["IndexCancer", "PreIndexCancer"], _3d=False,  row_min=1250, col_min=1250)
```

```python
# filter additionaly on the already filtered data
data.filter_data(age_range=[50, 70])
```

```python
age_list = []
for i in range(len(data)):
    if int(data[i].PatientAge[:-1]) >= (50*12) and int(data[i].PatientAge[:-1]) <= (70*12):
        age_list.append(data[i].PatientAge)
```

```python
len(age_list) # verifying all the ages in the data obj are indeed as specified by the filter
```
