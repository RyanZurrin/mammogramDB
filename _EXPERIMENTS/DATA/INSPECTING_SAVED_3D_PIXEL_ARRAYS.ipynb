```python
import numpy as np
import matplotlib.pyplot as plt

# Load your data
npz_data = np.load("/raid/mpsych/OMAMA/DATA/data/3d/BT.2.25.112767524153647278953555153211757459077.npz")
data = npz_data["data"]

# Choose a slice index
slice_index = 30  # Choose this number based on your specific needs

# Display that slice
plt.imshow(data[slice_index, :, :], cmap="gray")
plt.colorbar()
plt.title(f"Slice {slice_index}")
plt.show()
```

```python
from ipywidgets import interact

def show_slice(slice_idx):
    plt.imshow(data[slice_idx, :, :], cmap="gray")
    plt.colorbar()
    plt.title(f"Slice {slice_idx}")
    plt.show()

interact(show_slice, slice_idx=(0, data.shape[0]-1))
```
