```python
# Data
models = ['Baseline', 'GP2 Test 1', 'GP2 Test 2', 'GP2 Test 3']
train_dice_coeffs = [0.5494, 0.9623, 0.9486, 0.9725]
val_dice_coeffs = [0.5446, 0.8764, 0.8647, 0.8695]

# Bar chart
bar_width = 0.35
r1 = range(len(models))
r2 = [x + bar_width for x in r1]

plt.bar(r1, train_dice_coeffs, width=bar_width, color='b', align='center', label='Train Dice Coeff')
plt.bar(r2, val_dice_coeffs, width=bar_width, color='r', align='center', label='Validation Dice Coeff')

# Adding labels
plt.xlabel('Models', fontweight='bold')
plt.ylabel('Dice Coefficient', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(models))], models)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
models = ['Baseline', 'GP2 Test 1', 'GP2 Test 2', 'GP2 Test 3']
train_loss = [0.3997, 0.0360, 0.0501, 0.0263]
val_loss = [0.4099, 0.3247, 0.2542, 0.3957]

# Bar chart
bar_width = 0.35
r1 = range(len(models))
r2 = [x + bar_width for x in r1]

plt.bar(r1, train_loss, width=bar_width, color='b', align='center', label='Train Loss')
plt.bar(r2, val_loss, width=bar_width, color='r', align='center', label='Validation Loss')

# Adding labels
plt.xlabel('Models', fontweight='bold')
plt.ylabel('Loss', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(models))], models)
plt.legend()

plt.tight_layout()
plt.show()
```
