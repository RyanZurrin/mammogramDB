```python
def generate_error_codes_dict(csv_path=None):
    """
    Generate a dictionary of error codes and their corresponding
    descriptions.
    """
    error_codes_dict = {}
    if csv_path is None:
        csv_path = r'../omama/acceptance_criteria.csv'
        with open(csv_path, 'r') as csv_file:  
            lines = csv_file.readlines()
            for line in lines:
                values = line.split(',')
                error_code = values[0]
                print(error_code)
                description = values[1]
                print(description)
                error_codes_dict[error_code] = description
                
```
