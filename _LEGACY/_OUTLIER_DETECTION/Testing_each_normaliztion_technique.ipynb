```python
imgs = O.DataHelper.get2D(N=10, config_num=2)
```

# down sampling to 100x100

```python
downsampled_imgs = O.Normalize.get_norm(pixel_list, norm_type='downsample', output_shape=(100, 100))
```

```python
O.DataHelper.view_grid(downsampled_imgs)
```
