```python
import sys
import time
sys.path.insert(0,'../..')
import omama as O
import numpy as np
import mahotas as mh
import matplotlib.pyplot as plt
```

```python
imgs = O.DataHelper.get2D(N=10, config_num=2)
```

```python
imgs_norm_list = []
for img in imgs:
    temp = img.pixels.astype(np.float32)
    temp /= temp.max()
    temp *= 255
    imgs_norm_list.append(temp)
```

```python
img_hist_list = []
for norm_img in imgs_norm_list:
    img_hist_list.append(mh.fullhistogram(norm_img.astype(np.uint8)))
```

```python
for i in range(len(img_hist_list)):
    plt.hist(img_hist_list[i], log=True)
    plt.pause(1)
```

```python
temp_images = []
i = 0
for img in imgs_norm_list:
    temp_images.append(img.copy())
for img in temp_images:    
    _ = plt.hist(img.ravel(), bins = 16, log=True)
    _ = plt.xlabel('Intensity Value')
    _ = plt.ylabel('Count')
    _ = plt.ylim(0.1, 10**9)
    _ = plt.xlim(0.1, 256)
    plt.show()
    plt.pause(1)
    

# for i in range(len(img_hist_list)):
#     ax = plt.hist(img_hist_list[i].ravel(), bins = 256, log=True)
```
