```python
import sys
import time
sys.path.insert(0,'../..')
import omama as O
import numpy as np
import mahotas as mh
import matplotlib.pyplot as plt
```

```python
imgs = O.DataHelper.get2D(N=10, config_num=2)
```

```python
imgs_norm_list = []
for img in imgs:
    temp = img.pixels.astype(np.float32)
    temp /= temp.max()
    temp *= 255
    imgs_norm_list.append(temp)
```

```python
for img in imgs_norm_list:
    print(img.max())
    print(img.min())
```

```python
img_hist_list = []
for norm_img in imgs_norm_list:
    img_hist_list.append(mh.fullhistogram(norm_img.astype(np.uint8)))
```

```python
for item in img_hist_list:
    print(item.min())
```

```python
for item in img_hist_list:
    print(item.max())
```

```python
for i in range(len(img_hist_list)):
    plt.hist(img_hist_list[i])
    plt.pause(1)
```
